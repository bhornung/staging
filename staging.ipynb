# Staging business leads

## Motivation

A business lead is a contact with whom a company seeks to develop a relationship. As time progresses this relationship advances through various stages. From a superficial exchange through surveying the needs of the contact to its becoming a customer or a partner. The contact is can drop out irreversibly at any point. During this journey various pieces information about the contact is collected. This information is in turn used to decide what kind of relationship is sought and facilitate the process to achieve it.

The scope of this document is to build a mathematical toy model of this process. This model seeks to formalise the relationship between various stages. In addition this model should suggest ways to predict and facilitate the movement of contacts between the desired stages. 

## Introduction

### Definitions

A _stage_ is a well described kind of relationship. In the following we assume that the stage of a contact is always known. We do not seek to determine the stage of a contact based on the information gathered about it either.

```python
nodes = {'LD' : 'lead',
         'QLD' : 'qualified lead',
         'NEG' : 'negotiation',
         'PRT' : 'partner',
         'CST' : 'customer',
         'DRP' : 'dropout'}
    
edges = {nodes['LD'] : [nodes['QLD'], nodes['DRP']],
         nodes['QLD'] : [nodes['NEG'], nodes['DRP']],
         nodes['NEG'] : [nodes['CST'], nodes['PRT'], nodes['DRP']]}

edge_gen = ((_k, __v) for _k, _v in edges.items() for __v in _v)

pipeline = gv.Digraph(engine = 'dot')
pipeline.edges(edge_gen)
pipeline
```
