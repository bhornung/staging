# Staging business leads

## Motivation

A business lead is a contact with whom a company seeks to develop a relationship. As time progresses this relationship advances through various stages. From a superficial exchange through surveying the needs of the contact to its becoming a customer or a partner. The contact is can drop out irreversibly at any point. During this journey various pieces information about the contact is collected. This information is in turn used to decide what kind of relationship is sought and facilitate the process to achieve it.

The scope of this document is to build a mathematical toy model of this process. This model seeks to formalise the relationship between various stages. In addition this model should suggest ways to predict and facilitate the movement of contacts between the desired stages. 

## Introduction

### Definitions

A _stage_ is a well described kind of relationship. In the following we assume that the stage of a contact is always known. We do not seek to determine the stage of a contact based on the information gathered about it either.

**Stage indices**: 

$\mathbb{I}_{S} = \{1, 2, ..., N_{S} \}$ a set of indices are called _stage indices_. The elements in $\mathbb{I}_{S}$ will index the stages.

**Pool**:

$\mathbb{P}_{i}:\,\, i \in \mathbb{I}_{S}$ are called the _pools_. $\mathbb{P}_{i}$ will be the pool of stage $i$. The elements of a pool represents the unprocessed leads.

**Processed pool**

$\mathbb{M}_{i}: \,\, i \in \mathbb{I}_{S}$ are called the processed pools. They will contain contacts for which extra information have been gathered.

**Stages**

$S_{i} = \left(\mathbb{P}_{i}, \mathbb{M}_{i}\right): \, \, i \in \mathbb{I}_{S}$ are called _stages_.

**Set of stages** 

$\mathbb{S} = S_{i}, \,\, i \in \mathbb{I}_{S}\}$ .

The pipeline, $Q$ (replace with the appropriate bovine excrete term) is then defined as a directed graph
$$
    Q = (V, E),\, \, \text{where} \, V = S, E \in S^{2}
$$

### Attributes

The attributes are the quantities that are used to characterise a contact. In fact, we are going to define contacts as collections -- tuples -- of attributes

**Attribute indices**

The number of attributes is denoted by $N_{A}$. This is the number of all categories of information that can be possibly collected during any possible journey throught the pipeline. For convenience we introduce the index set of attributes:

$ \mathbb{I}_{A} = \{1, 2, ... N_{A}\}$

**Attributes**

An attribute $\mathbb{A}_{i}$ is a nonempety set. It represents a certain king of information gathered abount the contact It can be an continous, nominal or categorical variable. 

$\mathbb{A}_{i} \neq \emptyset, \,\, i \in  \mathbb{I}_{A}$

**Contact**

A contact, $x$ can be represented by a tuple over the attributes:

$x \in \left(\mathbb{A}_{1}, \mathbb{A}_{2}, ..., \mathbb{A}_{N_{A}} \right) = \mathbb{A}$

It is advantageous to groups attributes together that are collected in a certain stage.

$\mathbb{I}_{Ai} \subset \mathbb{I}_{A}$ is the index set of attributes collected have been by reaching the processed pool of stage $i$.

### Flow

The general workflow consists of two steps _collection_ and _moving_.

#### Measurement

_Measurement_ is the act of collecting the missing attributes at a stage. The measurment, consists of moving a contact from a pool to the measured pool.
$$
 f_{(M,i)}: \mathbb{P}_{i} \rightarrow \mathbb{M}_{i}
$$
    
During measurement the $\mathbb{I}_{Ai} \setminus \mathbb{I}_{Ai-1}$ attributes are populated.
    
#### Transfer 

_Transfer_ is the act of adding a contact in a cerain stage to the pool of an other stage.

$$
 f_{(T,i,j)}: \mathbb{M}_{i} \rightarrow \mathbb{P}_{j}, \, \, j \in N(S_{i}).
$$

where $N(S_{i})$ is the neighbourhood of stage $i$.

## What can be done?

### Transfer probability

The notation $(i,j)$ refers to edges in $Q$ from now on.

The probability of transfer $i\rightarrow j$ is denoted by $p_{ij} = p_{ij}(x), \,\, x \in \mathbb{A}$. The main goal is characterise the probability transfer functions as a function of the attributes. These probabilities can be represented as a transfer matrix, $\mathbf{P}$.

$$
\mathbf{P} \in [0,1]^{N_{S} \times N_{S}}
$$

Construction of $\mathbf{P}$ as a function of $x$ in the vast majority of cases is not feasible (think about the dimensionality of the problem). Instead, we try to find regularities that can help to predict the value or behaviour of $\mathbf{P}$.
